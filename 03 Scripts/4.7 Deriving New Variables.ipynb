### Defining a function to create price labels

```python
# We are defining our function for creating price_label
def price_label(row):

  if row['prices'] <= 5:# Here we are selecting rows in the 'prices' column
    return 'Low-range product'# Anything below $5 will be labeled low range etc., else if (elif) are similar in structure
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Now we will use our defined function price_label,
# We are creating a new column, price_range and using our price_label function to fill the data from the prices column
df['price_range'] = df.apply(price_label, axis=1)
```

```python
# Checking the values inputed
df['price_range'].value_counts(dropna = False)
```

```python
# Checking max price in our subset
df['prices'].max()
```

### Using loc() function rather than defining our own funciton

```python
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product' 
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

##  Clients request to find busiest and least busiest days of week

```python
# Frequency of orders on what day of the week they occur
ords_prods_merged['orders_day_of_week'].value_counts(dropna = False)
```

```python
# Using a for-loop to fill our result list of frequencies of how busy a day of the week is
result = []

for value in ords_prods_merged["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
result
```

```python
# Creating a new column and setting our result list to fill it
ords_prods_merged['busiest_day'] = result
```

```python
# Checking value counts of busiest_day column
ords_prods_merged['busiest_day'].value_counts(dropna = False)
```

### Quesiton 4 - Senior Tech requests we find busiest times and slowest times of day for trouble-shooting

```python
#Finding value counts for order hours in given day
ords_prods_merged['order_hour_of_day'].value_counts()
```

```python
# Divided our 24 hour clock into 3 parts/ 8 hours per part

hour_result = []

for value in ords_prods_merged['order_hour_of_day']:
    if value == 10 or value == 11 or value == 14 or value == 15 or value == 13 or value == 12 or value == 16 or value == 9:
      hour_result.append('Most Orders')
    elif value == 3 or value == 4 or value == 2 or value == 5 or value == 1 or value == 0 or value == 6 or value == 23:
      hour_result.append('Average Orders')
    else:
      hour_result.append('Fewest Orders')
```

```python
ords_prods_merged['busiest_period_of_day'] = hour_result
```
